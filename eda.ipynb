```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_labels = np.load('data/train/train_labels.npy')
train_seismic = np.load('data/train/train_seismic.npy')
```

```python
train_labels.shape, train_seismic.shape
```

```python
# Example: Plot a depth slice at index 100
depth_index = 123
plt.subplot(121)
plt.imshow(train_seismic[:, :, depth_index], cmap='seismic', aspect='auto')
plt.colorbar(label='Amplitude')
plt.title(f"Depth Slice at Index {depth_index}")
plt.xlabel("Crossline")
plt.ylabel("Inline")
plt.subplot(122)
plt.imshow(train_labels[:, :, depth_index], cmap='seismic', aspect='auto')
plt.colorbar(label='Amplitude')
plt.xlabel("Crossline")
plt.ylabel("Inline")
plt.tight_layout()
plt.show()
```

```python
def plot_5_slices(type):
    fig, axs = plt.subplots(2, 5, figsize=(20, 4),sharex=True, sharey=True)
    for i in range(5):
        depth_index = i * 50
        if type == 'inline':
            slice_data = train_seismic[depth_index, :, :].T
            label_data = train_labels[depth_index, :, :].T
        elif type == 'crossline':
            slice_data = train_seismic[:, depth_index, :].T
            label_data = train_labels[:, depth_index, :].T
        elif type == 'depth':
            slice_data = train_seismic[:, :, depth_index]
            label_data = train_labels[:, :, depth_index]
        axs[0,i].imshow(slice_data, cmap='seismic', aspect='auto')
        axs[0,i].set_title(f"{type.capitalize()} Slice at Index {depth_index}")
        axs[0,i].set_xlabel("Crossline")
        axs[0,i].set_ylabel("Inline")
        axs[1,i].imshow(label_data, cmap='seismic', aspect='auto')
        axs[1,i].set_xlabel("Crossline")
        axs[1,i].set_ylabel("Inline")
    plt.tight_layout()
    plt.show()
```

```python
plot_5_slices('inline')
plot_5_slices('crossline')
plot_5_slices('depth')
```
